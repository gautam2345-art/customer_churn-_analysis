To make the data easy to read, we are using a function to change the demonstration of df "SeniorCitizen" from

```python
def conv(value):

  if value==1:
    return "Yes"
  else:
    return "No"
df['SeniorCitizen'] = df['SeniorCitizen'] .apply(lambda x: conv(x))
```

This visualization helps in understanding how different services and contract types relate to customer churn.



```python
#list of categorical column
columns = ['PhoneService', 'MultipleLines', 'InternetService',
           'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
           'TechSupport', 'StreamingTV', 'StreamingMovies']

#define subplot layout(3 rows, 3 column)
fig, axes= plt.subplots(3,3,figsize=(15,10))
fig.suptitle("Countplots for Categorical Features", fontsize=16)

# Loop through columns and create countplots in subplots
for i, col in enumerate(columns):
    row, col_index = divmod(i, 3)  # Determine row and column position
    sns.countplot(x=df[col], ax=axes[row, col_index], palette="Set1", hue = df["Churn"])
    axes[row, col_index].set_title(f"Countplot of {col}")  # Set title
    axes[row, col_index].set_xlabel("")  # Remove x-label for cleaner look
    axes[row, col_index].set_ylabel("Count")  # Label y-axis

# Adjust layout
plt.tight_layout(rect=[0, 0, 1, 0.96])  # Prevent title from overlapping
plt.show()


```
